```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
fare = df[['Fare']].dropna()
```

```python
plt.figure(figsize=(5,3))
sns.histplot(fare['Fare'], kde=True)
plt.title("Original Fare Distribution")
plt.show()
```

```python
fare['Fare_log'] = np.log1p(fare['Fare'])


```

```python
plt.figure(figsize=(5,3))
sns.histplot(fare['Fare_log'], kde=True)
plt.title("Log Transformed Fare")
plt.show()
```

```python
#When to use? Moderate skewness ,Count-type data

fare['Fare_sqrt'] = np.sqrt(fare['Fare'])

plt.figure(figsize=(5,3))
sns.histplot(fare['Fare_sqrt'], kde=True)
plt.title("Sqrt Transformed Fare")
plt.show()
```

```python
#When to use? Very strong skew Outliers dominate
fare['Fare_reciprocal'] = 1 / fare['Fare']

plt.figure(figsize=(5,3))
sns.histplot(fare['Fare_reciprocal'], kde=True)
plt.title("Reciprocal Transformed Fare")
plt.show()
```
