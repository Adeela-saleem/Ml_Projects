## Handling Outliers

```python
cols = ['AveRooms', 'AveBedrms', 'Population', 'AveOccup', 'MedInc']
```

```python
def detect_outliers_IQR(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    print(f"{col} -> {len(outliers)} outliers")
    return lower_bound, upper_bound

for col in cols:
    lb, ub = detect_outliers_IQR(X_train, col)
```

```python
for col in cols:
    lb, ub = detect_outliers_IQR(X_train, col)
    X_train[col] = X_train[col].clip(lb, ub)
    X_test[col] = X_test[col].clip(lb, ub)
```

```python
for col in cols:
    plt.figure(figsize=(12,5))

    # Before clip
    plt.subplot(1,2,1)
    sns.boxplot(x=X_train[col], color='skyblue')
    plt.title(f"Boxplot Before - {col}")

    # After clip (agar clip kiya)
    plt.subplot(1,2,2)
    sns.boxplot(x=X_train[col], color='lightgreen')
    plt.title(f"Boxplot After - {col}")

    plt.show()
```

### Scaling

```python
from sklearn.preprocessing import StandardScaler

# Skewed / transformed columns
cols_to_scale = ['MedInc', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup']

# Create scaler object
scaler = StandardScaler()
```

```python
# Fit scaler on train
X_train_scaled = X_train.copy()
X_train_scaled[cols_to_scale] = scaler.fit_transform(X_train[cols_to_scale])

# Transform test using train parameters
X_test_scaled = X_test.copy()
X_test_scaled[cols_to_scale] = scaler.transform(X_test[cols_to_scale])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in cols_to_scale:
    plt.figure(figsize=(12,4))

    plt.subplot(1,2,1)
    sns.histplot(X_train[col], kde=True, color='skyblue')
    plt.title(f"Before Scaling - {col}")

    plt.subplot(1,2,2)
    sns.histplot(X_train_scaled[col], kde=True, color='salmon')
    plt.title(f"After Scaling - {col}")

    plt.show()
```

```python
# Agar log-transform apply karna hai
y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Model object create
lr = LinearRegression()

# Train model
lr.fit(X_train, y_train_log)   # ya y_train without log

# Predict
y_pred_log = lr.predict(X_test)

# Inverse log-transform to original scale
y_pred = np.expm1(y_pred_log)
```

```python
# Metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"RMSE: {rmse}")
print(f"R² Score: {r2}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.scatterplot(x=y_test, y=y_pred)
plt.xlabel("Actual MedHouseVal")
plt.ylabel("Predicted MedHouseVal")
plt.title("Actual vs Predicted - Linear Regression")
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()], 'r--')  # ideal line
plt.show()
```

```python
residuals = y_test - y_pred

plt.figure(figsize=(8,6))
sns.scatterplot(x=y_pred, y=residuals)
plt.axhline(0, color='r', linestyle='--')
plt.xlabel("Predicted MedHouseVal")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```
